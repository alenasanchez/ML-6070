## Import Packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import KNeighborsClassifier
```

## Read Data

```python
url = "https://github.com/ArinB/CA05-kNN/raw/master/movies_recommendation_data.csv"
data = pd.read_csv(url, encoding = "ISO-8859-1") 
```

### clean the data and prep for algorithm

```python
data = data.drop(labels = ["Label", "Movie ID"], axis = 1)
```

```python
data.head()
```

### split the data

```python
x = data.loc[:, data.columns != 'Movie Name']
y = data.loc[:, data.columns == 'Movie Name']
```

### check to ensure split is correct

```python
x.head()
```

```python
y.head()
```

## Recommender System

```python
#Create Recommender System using KNeighborsClassifier from skicit learn and set n_neighbors hyperparameter to 5!
neighbors = KNeighborsClassifier(n_neighbors=5)
#fit the model
neighbors.fit(x, y)
```

```python
#Determine Nearest Neighbors for The Post using the genre information provided in the assignmnet 
#IMDB Rating = 7.2, Biography = Yes, Drama = Yes, Thriller = No, Comedy = No, Crime = No, Mystery = No, History = Yes

post_neightbors = neighbors.kneighbors([[7.2,1,1,0,0,0,0,1]])
post_neightbors
```

```python
#Use a for loop to print the titles of the 5 movies given from the algorithm.
print("Here are 5 similar movies to The Post: ")
print("----------------------------------------")
for i in post_neightbors[1][0]:
  print(data.loc[i]['Movie Name'])
```
